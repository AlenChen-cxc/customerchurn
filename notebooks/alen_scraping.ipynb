```python
import requests
from bs4 import BeautifulSoup

response = requests.get('https://www.productreview.com.au/listings/seniors-car-insurance?page=1#reviews')
soup = BeautifulSoup(response.content, "html.parser")

box_class = 'p-4_xQO card-body_d_t flex-column__BR d-flex_RNf stackY-4_EQi'
stars_class = 'me-1_KrM flex-shrink-0_YwL rating_A5M rating--dark_N9b rating--40__ae rating--medium__tq rating--stars_CTa'
text_class= 'mb-0__MF'

reviews = []

for review in soup.find_all('div', class_=box_class):
    stars = review.find('div', class_=stars_class).title.string
    text = review.find('p', class_=text_class)
    reviews.append({'stars': stars, 'review': text})
```
